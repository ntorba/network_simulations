```python
def cyto_converter(G):
    cyto_nodes = []
    cyto_edges = []
    for node in G.nodes: 
        current = {
            'data':{'id': node, 'label': G.nodes[node].get("label", f"label-{node}")}
        }
        cyto_nodes.append(current)

    for n1, n2 in G.edges: 
        current = {'data': {'source': n1, 'target': n2}}
        cyto_edges.append(current)
    
    return cyto_nodes, cyto_edges 
nodes, edges = cyto_converter(G)
```

```python
import dash
import dash_cytoscape as cyto
import dash_html_components as html

app = dash.Dash(__name__)

app.layout = html.Div([
    cyto.Cytoscape(
        id='cytoscape-two-nodes',
        layout={'name': 'preset'},
        style={'width': '100%', 'height': '400px'},
        elements=nodes+edges
    )
])
```
